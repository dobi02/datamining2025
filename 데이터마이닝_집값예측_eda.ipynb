```python
# Í≤∞Ï∏°Ïπò ÎπÑÏú®Ïù¥ ÎÜíÏùÄ Ïó¥ÏùÑ ÏÇ≠Ï†ú (ÏûÑÍ≥ÑÍ∞í ÏÑ§Ï†ï: 30% Ïù¥ÏÉÅ Í≤∞Ï∏°Ïù¥Î©¥ ÏÇ≠Ï†ú)
threshold = 0.3  # 30%
missing_ratio = df.isnull().mean()
columns_to_drop = missing_ratio[missing_ratio > threshold].index

# ÏÇ≠Ï†úÌï† Ïó¥ Î™©Î°ù ÌôïÏù∏
columns_to_drop_list = list(columns_to_drop)

# ÏõêÎ≥∏ Îç∞Ïù¥ÌÑ∞ÏóêÏÑú Ïó¥ ÏÇ≠Ï†ú
df_cleaned = df.drop(columns=columns_to_drop_list)

import pandas as pd
import matplotlib.pyplot as plt

# ÏÇ≠Ï†ú ÌõÑ ÎÇ®ÏùÄ Ïó¥ Ïàò
remaining_columns = df_cleaned.shape[1]

# Í≤∞Ï∏°ÏπòÍ∞Ä Ïó¨Ï†ÑÌûà ÎÇ®ÏïÑ ÏûàÎäî Ïó¥ Í∞úÏàò ÌôïÏù∏
remaining_missing = df_cleaned.isnull().sum()
remaining_missing = remaining_missing[remaining_missing > 0]

remaining_columns, columns_to_drop_list, remaining_missing.head()
```

```python
# ÏàòÏπòÌòï Î≥ÄÏàò ‚Üí Ï§ëÏïôÍ∞íÏúºÎ°ú Í≤∞Ï∏°Ïπò Ï±ÑÏö∞Í∏∞
num_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns
df_cleaned[num_cols] = df_cleaned[num_cols].fillna(df_cleaned[num_cols].median())

# Î≤îÏ£ºÌòï Î≥ÄÏàò ‚Üí ÏµúÎπàÍ∞íÏúºÎ°ú Í≤∞Ï∏°Ïπò Ï±ÑÏö∞Í∏∞
cat_cols = df_cleaned.select_dtypes(include=['object']).columns
df_cleaned[cat_cols] = df_cleaned[cat_cols].fillna(df_cleaned[cat_cols].mode().iloc[0])

# Í≤∞Ï∏°Ïπò Ï≤òÎ¶¨ ÏôÑÎ£å ÌõÑ ÌôïÏù∏
missing_after = df_cleaned.isnull().sum().sum()
print(f"ÎÇ®ÏùÄ Í≤∞Ï∏°Ïπò Ïàò: {missing_after}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ÏàòÏπòÌòï Î≥ÄÏàòÎßå ÏÑ†ÌÉù
num_cols = df_cleaned.select_dtypes(include=['float64', 'int64']).columns

# Ïù¥ÏÉÅÏπò Ï†úÍ±∞Ìï† Îç∞Ïù¥ÌÑ∞ Î≥µÏÇ¨
df_outlier_removed = df_cleaned.copy()

# Í∑∏ÎûòÌîÑÏóê Í∑∏Î¶¥ Î≥ÄÏàò Ïàò Ï†úÌïú (ÎÑàÎ¨¥ ÎßéÏúºÎ©¥ ÌûòÎìúÎãàÍπå Top 6Îßå ÏòàÏãú)
plot_cols = num_cols[:6]

# üîç Ïù¥ÏÉÅÏπò ÏãúÍ∞ÅÌôî + Ï†úÍ±∞ (IQR Î∞©Ïãù)
for col in plot_cols:
    # Boxplot Í∑∏Î¶¨Í∏∞ (Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†Ñ)
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_outlier_removed[col])
    plt.title(f"üì¶ Boxplot Before Outlier Removal - {col}")
    plt.grid(True)
    plt.show()

    # IQR Í≥ÑÏÇ∞
    Q1 = df_outlier_removed[col].quantile(0.25)
    Q3 = df_outlier_removed[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    # Ïù¥ÏÉÅÏπò Ï†úÍ±∞
    df_outlier_removed = df_outlier_removed[(df_outlier_removed[col] >= lower) & (df_outlier_removed[col] <= upper)]

    # Boxplot Í∑∏Î¶¨Í∏∞ (Ïù¥ÏÉÅÏπò Ï†úÍ±∞ ÌõÑ)
    plt.figure(figsize=(6, 4))
    sns.boxplot(x=df_outlier_removed[col])
    plt.title(f"‚úÖ Boxplot After Outlier Removal - {col}")
    plt.grid(True)
    plt.show()

print(f"üéØ Ïù¥ÏÉÅÏπò Ï†úÍ±∞ ÌõÑ Îç∞Ïù¥ÌÑ∞ ÌÅ¨Í∏∞: {df_outlier_removed.shape}")
```

```python
# SalePrice Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†ÑÌõÑ Î∂ÑÌè¨ ÎπÑÍµê
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
sns.histplot(df_cleaned['SalePrice'], kde=True)
plt.title("SalePrice Î∂ÑÌè¨ (Before)")

plt.subplot(1, 2, 2)
sns.histplot(df_outlier_removed['SalePrice'], kde=True)
plt.title("SalePrice Î∂ÑÌè¨ (After)")
plt.tight_layout()
plt.show()

# Í∏∞Ïà†ÌÜµÍ≥Ñ ÎπÑÍµê
print("Before Outlier Removal:\n", df_cleaned['SalePrice'].describe())
print("\nAfter Outlier Removal:\n", df_outlier_removed['SalePrice'].describe())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# üìå 4. SalePrice Î∂ÑÌè¨ ÌôïÏù∏ (Î°úÍ∑∏ Ï†Ñ/ÌõÑ)
plt.figure(figsize=(6, 4))
sns.histplot(df_outlier_removed['SalePrice'], kde=True)
plt.title("SalePrice Î∂ÑÌè¨ (Î°úÍ∑∏ Î≥ÄÌôò Ï†Ñ)")
plt.show()

plt.figure(figsize=(6, 4))
sns.histplot(np.log1p(df_outlier_removed['SalePrice']), kde=True)
plt.title("SalePrice Î∂ÑÌè¨ (Î°úÍ∑∏ Î≥ÄÌôò ÌõÑ)")
plt.show()
```

```python
# 4Îã®Í≥Ñ: Î≥ÄÏàò Í∞Ñ Í¥ÄÍ≥Ñ Î∂ÑÏÑù

# Î°úÍ∑∏ Î≥ÄÌôòÎêú SalePrice ÏÉùÏÑ±
df_outlier_removed['SalePrice_log'] = np.log1p(df_outlier_removed['SalePrice'])

# ÏàòÏπòÌòï Î≥ÄÏàòÏôÄ Î°úÍ∑∏ SalePrice Í∞Ñ ÏÉÅÍ¥ÄÍ≥ÑÏàò Í≥ÑÏÇ∞
correlations = df_outlier_removed.corr(numeric_only=True)['SalePrice_log'].sort_values(ascending=False)

# ÏÉÅÏúÑ 10Í∞ú ÏÉÅÍ¥Ä Î≥ÄÏàò Ï∂îÏ∂ú
top_corr_features = correlations[1:11]  # SalePrice_log Ï†úÏô∏
top_corr_features
```

```python
# 4Îã®Í≥Ñ Í≥ÑÏÜç: ÏÉÅÏúÑ ÏÉÅÍ¥Ä Î≥ÄÏàòÎì§Ïùò ÏÇ∞Ï†êÎèÑ ÏãúÍ∞ÅÌôî

top_features = top_corr_features.index

# ÏÇ∞Ï†êÎèÑ Í∑∏Î¶¨Í∏∞
for feature in top_features:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=df_outlier_removed[feature], y=df_outlier_removed['SalePrice_log'])
    plt.title(f"{feature} vs SalePrice_log")
    plt.xlabel(feature)
    plt.ylabel("SalePrice (log)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# ÌûàÌä∏Îßµ: ÏÉÅÏúÑ ÏÉÅÍ¥Ä Î≥ÄÏàòÎì§ Í∞Ñ ÏÉÅÍ¥ÄÍ¥ÄÍ≥Ñ
plt.figure(figsize=(10, 8))
sns.heatmap(df_outlier_removed[top_features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title("ÏÉÅÏúÑ ÏÉÅÍ¥Ä Î≥ÄÏàò Í∞Ñ ÌûàÌä∏Îßµ")
plt.tight_layout()
plt.show()
```

```python
# 5Îã®Í≥Ñ: Í∑∏Î£πÎ≥Ñ Î∂ÑÏÑù - Î≤îÏ£ºÌòï Î≥ÄÏàò Í∏∞Ï§Ä Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† SalePrice (Î°úÍ∑∏) + Î∞ïÏä§ÌîåÎ°Ø

# ÏÇ¨Ïö©Ìï† Ï£ºÏöî Î≤îÏ£ºÌòï Î≥ÄÏàò Î™©Î°ù
group_vars = ['Neighborhood', 'House Style', 'Overall Qual']

# Î°úÍ∑∏ Î≥ÄÌôòÌïú SalePrice Ï∂îÍ∞Ä
df_outlier_removed['SalePrice_log'] = np.log1p(df_outlier_removed['SalePrice'])

# Í∞Å Î≤îÏ£ºÌòï Î≥ÄÏàòÎ≥ÑÎ°ú ÌèâÍ∑† SalePrice_log Í≥ÑÏÇ∞ Î∞è ÏãúÍ∞ÅÌôî
for var in group_vars:
    # Í∑∏Î£πÎ≥Ñ ÌèâÍ∑†
    group_means = df_outlier_removed.groupby(var)['SalePrice_log'].mean().sort_values(ascending=False)

    # ÎßâÎåÄÍ∑∏ÎûòÌîÑ
    plt.figure(figsize=(10, 5))
    sns.barplot(x=group_means.index, y=group_means.values)
    plt.xticks(rotation=45)
    plt.title(f"Í∑∏Î£πÎ≥Ñ ÌèâÍ∑† SalePrice_log - {var}")
    plt.ylabel("ÌèâÍ∑† SalePrice (log)")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Î∞ïÏä§ÌîåÎ°Ø
    plt.figure(figsize=(10, 5))
    sns.boxplot(x=var, y='SalePrice_log', data=df_outlier_removed)
    plt.xticks(rotation=45)
    plt.title(f"Î∞ïÏä§ÌîåÎ°Ø: SalePrice_log by {var}")
    plt.ylabel("SalePrice (log)")
    plt.tight_layout()
    plt.show()
```
